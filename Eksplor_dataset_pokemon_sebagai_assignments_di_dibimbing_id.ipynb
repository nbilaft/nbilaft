## mengjitung korlasi



*   df[['Health Points', 'Attack Points', 'Defense Points', 'Special Attack', 'Special Defense', 'Speed Points']]: Ini adalah cara untuk memilih kolom-kolom tertentu dari DataFrame df. Dalam kasus ini, kita memilih kolom-kolom 'Health Points', 'Attack Points', 'Defense Points', 'Special Attack', 'Special Defense', dan 'Speed Points'.

*   corr(numeric_only=True): Metode corr() digunakan untuk menghitung matriks korelasi antar kolom-kolom yang dipilih. Parameter numeric_only=True memastikan bahwa hanya kolom-kolom dengan tipe data numerik yang dihitung korelasinya.

# Membuat Heatmap


*   sns.heatmap(): Ini adalah fungsi dari seaborn (sns) yang digunakan untuk membuat heatmap (peta panas). Heatmap ini memvisualisasikan data dalam bentuk matriks dengan menggunakan warna untuk mewakili nilai-nilai di dalamnya.

*   corr: Variabel corr yang merupakan hasil dari perhitungan korelasi sebelumnya, digunakan sebagai input untuk heatmap. Ini berarti heatmap akan menampilkan korelasi antar kolom-kolom yang telah dipilih.

*   annot=True: Parameter ini membuat heatmap menampilkan nilai-nilai yang sesuai di dalam setiap selnya. Ini membantu untuk dengan cepat melihat nilai korelasi yang terdapat dalam heatmap.

*   cmap='coolwarm': Parameter ini mengatur skema warna yang digunakan untuk heatmap. Dalam contoh ini, digunakan skema warna coolwarm yang berbeda-beda antara warna dingin dan hangat untuk menunjukkan korelasi positif dan negatif.

*   fmt=".2f": Parameter ini mengatur format dari nilai-nilai yang ditampilkan di dalam setiap sel heatmap. Di sini, nilai-nilai ditampilkan dengan dua desimal angka belakang koma untuk korelasi.










```
corr = df[['Health Points', 'Attack Points', 'Defense Points', 'Special Attack', 'Special Defense', 'Speed Points']].corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
```
